```python
# Importing data
import pandas as pd
dataset = pd.read_csv("D:\Bitan\IIAS\python_project.csv")
print(dataset)
```

```python
# Connecting to the MySQL and creating a database
import mysql.connector as msql
from mysql.connector import Error
try:
    conn = msql.connect(host='localhost', user='root',  
                        password='bitan27', port='3308')
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("CREATE DATABASE irisDB")
        print("irisDB database is created")
except Error as e:
    print("Error while connecting to MySQL", e)
```

```python
# Getting the data types
dataset.dtypes
```

```python
# Framing the query
query = '''CREATE TABLE iris(open FLOAT(5,2),close FLOAT(5,2),low FLOAT(5,2),
           high FLOAT(5,2))'''
# Creating a table & Import the CSV data into the MySQL table
try:
    conn = msql.connect(host='localhost', 
                           database='irisDB', user='root', 
                           password='bitan27', port='3308')
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        cursor.execute('DROP TABLE IF EXISTS iris;')
        print('Creating table....')
        cursor.execute(query)
        print("iris table is created....")
        for i,row in dataset.iterrows():
            sql = "INSERT INTO iris VALUES (%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")
            # the connection is not autocommitted by default, so we must commit to save our changes
            conn.commit()
except Error as e:
    print("Error while connecting to MySQL", e)
```

```python
#Query the database to check our work
# Execute query
sql = '''SELECT COUNT(open) 
         FROM `irisdb`.`iris`;'''
cursor.execute(sql)

# Count the number of rows inserted
result = cursor.fetchall()
for i in result:
    print(i)
```

```python
# Execute query
sql = "SELECT * FROM iris"
cursor.execute(sql)

# Fetch all the records
result = cursor.fetchall()
for i in result:
    print(i)
```
